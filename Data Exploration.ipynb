```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('Finstatements.csv')
```

```python
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
```

```python
df
```

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow from Operations Growth(%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
df.to_csv('Finstatements_Final.csv')
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth (%)' : 'mean',
    'Liabilities Growth (%)' : 'mean',
    'Cash Flow from Operations Growth(%)' :'mean'    
}).reset_index()
```

```python
summary
```

# Observations and Analysis:

## Company Performance:

1. Tesla shows the highest average growth rates across most financial metrics, indicating strong performance and significant expansion over the period.
2. Microsoft exhibits moderate growth rates, with consistent positive trends across all metrics.
3. Apple's growth rates are comparatively lower than Tesla and Microsoft, suggesting relatively stable performance.


## Revenue Growth:

1. Tesla demonstrates exceptional revenue growth, averaging over 23% per year, which could be attributed to increased demand for electric vehicles and expansion into new markets.
2. Microsoft maintains steady revenue growth, averaging around 8%, reflecting the company's successful product launches and strategic acquisitions.
3. Apple's revenue growth, while positive, is relatively modest compared to Tesla and Microsoft, possibly due to market saturation and slower product innovation.

## Net Income Growth:

1. Tesla's net income growth is remarkable, averaging around 47% per year, indicating efficient cost management and strong profitability.
2. Microsoft and Apple also show positive net income growth, albeit at lower rates compared to Tesla, reflecting their solid financial performance and profitability.

## Asset and Liability Growth:

1. Tesla's rapid asset and liability growth, averaging around 20% and 12% per year respectively, signify aggressive expansion and investment activities.
2. Microsoft and Apple exhibit more moderate growth rates in assets and liabilities, indicating a more balanced approach to business expansion and financial management.

## Cash Flow from Operations:

1. Tesla's cash flow from operations growth, while positive, shows some volatility, possibly due to the capital-intensive nature of its business operations.
2. Microsoft and Apple demonstrate relatively stable cash flow growth, reflecting their strong operational efficiency and cash generation capabilities.

# Conclusions:

1. Tesla's Growth Trajectory: Tesla emerges as the standout performer with exceptional growth rates across all financial metrics, driven by its innovative products and aggressive expansion strategies.
2. Stability of Microsoft and Apple: Microsoft and Apple exhibit more stable growth patterns, reflecting their established positions in the market and robust business models.
3. Investment Considerations: Investors may find Tesla attractive for its high growth potential, but should also consider its higher risk profile compared to more established companies like Microsoft and Apple.


# Additional Analysis and Viz

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Total Revenue', hue='Company', marker='o')
plt.title('Total Revenue Over Time')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.legend(title='Company')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Year', y='Total Revenue', hue='Company')
plt.title('Total Revenue Comparison Across Companies')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.legend(title='Company')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Revenue Growth (%)', hue='Company', marker='o')
plt.title('Revenue Growth Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Revenue Growth Rate (%)')
plt.grid(True)
plt.legend(title='Company')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Total Revenue', bins=20, kde=True)
plt.title('Distribution of Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Company', y='Total Revenue')
plt.title('Total Revenue Distribution Across Companies')
plt.xlabel('Company')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Total Assets', y='Total Liabilities', hue='Company')
plt.title('Total Assets vs Total Liabilities')
plt.xlabel('Total Assets')
plt.ylabel('Total Liabilities')
plt.grid(True)
plt.legend(title='Company')
plt.show()
```
